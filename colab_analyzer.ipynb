```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_df  = pd.read_csv("rnet_training/logs/train_log.csv")
val_df  = pd.read_csv("rnet_training/logs/val_log.csv")
```

```python
def plot_progress(train_df, val_df, title=""):
    fig, axis = plt.subplots(ncols=3, figsize=(15, 4))
    fig.suptitle(title)
    axis[0].plot(train_df.total_loss, label="train")
    axis[0].plot(val_df.total_loss, label="val")
    axis[0].set_title("total_loss")
    axis[1].plot(train_df.detection_loss, label="train")
    axis[1].plot(val_df.detection_loss, label="val")
    axis[1].set_title("detection_loss")
    axis[2].plot(train_df.bbox_loss, label="train")
    axis[2].plot(val_df.bbox_loss, label="val")
    axis[2].set_title("bbox_loss")
    for i in range(3):
        axis[i].grid()
        axis[i].legend()
        axis[i].set_xlabel("epoch")
        axis[i].set_ylabel("loss")
    plt.show()
```

```python
plot_progress(train_df, val_df, title="Train/Val")
```
